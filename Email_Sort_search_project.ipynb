```python
#
import random
import string
```

```python
# list of emails domain
domain=['gmail.com','yahoo.com','aol.com','outlook.com','inmail.com','japan.com','usmail.com']
```

```python
# generate random email with random char and digit and random domain
chars = string.ascii_letters
digits = string.digits
def gen_emails(n,domain):
    email_list = []
    for d in domain:
        email_list.append(''.join(random.choice(chars) for _ in range(15))+'_'+''.join(random.choice(digits) for _ in range(2))+'@'+d)
    return email_list
```

```python
# generate email and write to a file
f = open("/home/sandipan/Documents/email_list.txt", "a")
for x in range(1000):
    f.write('\n'.join(gen_emails(12,domain)))
    f.write('\n')
f.close
```
